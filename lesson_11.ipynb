Для f(x)  = -5*x**2 + 10*x - 150 
1. Построить график

```python
from sympy import *
from sympy.plotting import plot
init_printing()
import matplotlib.pyplot as plt
import math
import matplotlib
import sympy
from sympy.plotting import plot
sympy.init_printing() 
```

```python
x = Symbol('x')
f = -5*x**2 + 10*x - 150
f
```

```python
x = Symbol('x')
plot(f, (x, -15, 15))# график на отрезке
```

2. Найти интервалы, на которых функция возрастает

```python
solve(diff(f)>0)
```

3. Найти интервалы, на которых функция убывает

```python
solve(diff(f)<0)
```

4. Определить корни

```python
solve(f)
#x1, x2 = solve(fx)
#print('корень x1: ', x1.evalf(), '\nкорень x2: ', round(x2.evalf(),2) )
```

5. Вычислить вершину

```python
x=solve(diff(f))
print('x = ', x )
```

```python
f.subs(x, 1)
#print('y = ', -5*1**2 + 10*1 - 150)
```

6. Определить промежутки, на котором f > 0

```python
print(solve(f>0))
```

7. Определить промежутки, на котором f < 0

```python
print(solve(f<0))
```
